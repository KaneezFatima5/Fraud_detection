```python
python
import pandas as pd
import numpy as np

# --- Dummy Data for Demonstration ---
# In your actual notebook, load your real data here
data = pd.DataFrame({
    'ProductCD': ['A', 'B', 'A', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'card4': ['visa', 'master', 'visa', 'visa', 'master', 'visa', 'visa', 'master', 'visa', 'master'],
    'isFraud': [0, 1, 0, 0, 1, 0, 1, 0, 1, 1]
})


# --- TreeNode Class ---
class TreeNode:
    def __init__(self, feature_index=None, split_value=None, children=None, leaf_value=None):
        self.feature_index = feature_index  # The index of the feature to split on
        self.split_value = split_value      # The value of the feature for the split
        self.children = children or {}      # A dictionary of children nodes {value: child_node}
        self.leaf_value = leaf_value        # The predicted class for a leaf node

    def print_tree(self, level=0):
        """Prints the tree structure for visualization."""
        prefix = '  ' * level
        if self.leaf_value is not None:
            print(f"{prefix}-- Leaf: Predict {self.leaf_value}")
        else:
            print(f"{prefix}-- Split on feature {self.feature_index} with value {self.split_value}")
            for value, child in self.children.items():
                print(f"{prefix}    -- Branch for value: {value}")
                child.print_tree(level + 1)


# --- Impurity Functions (Gini for demonstration) ---
def gini_impurity(y):
    """Calculates the Gini Impurity for a list of class labels."""
    if len(y) == 0:
        return 0
    probabilities = [np.sum(y == c) / len(y) for c in np.unique(y)]
    return 1 - sum(p**2 for p in probabilities)

def information_gain(data, split_feature, target_feature):
    """Calculates Information Gain using Gini Impurity."""
    parent_impurity = gini_impurity(data[target_feature])
    weighted_child_impurity = 0
    unique_values = data[split_feature].unique()
    
    for value in unique_values:
        subset = data[data[split_feature] == value]
        weighted_child_impurity += (len(subset) / len(data)) * gini_impurity(subset[target_feature])
        
    return parent_impurity - weighted_child_impurity


# --- Tree Building Functions ---
def find_best_split(data, features, target_feature):
    """Finds the best feature to split the data on based on Information Gain."""
    best_gain = -1
    best_feature = None
    
    for feature in features:
        gain = information_gain(data, feature, target_feature)
        if gain > best_gain:
            best_gain = gain
            best_feature = feature
            
    return best_feature, best_gain

def create_tree(data, features, target_feature, max_depth, current_depth=0):
    """
    Recursively builds a decision tree.
    
    This function will serve as the core of your individual decision tree
    for a random forest.
    """
    # Check stopping conditions
    if current_depth >= max_depth or len(data[target_feature].unique()) == 1 or len(data) < 2:
        leaf_value = data[target_feature].mode()[0] if not data.empty else None
        return TreeNode(leaf_value=leaf_value)

    # Find the best feature to split on
    best_feature, best_gain = find_best_split(data, features, target_feature)
    if best_gain <= 0:  # No further gain, stop splitting
        leaf_value = data[target_feature].mode()[0]
        return TreeNode(leaf_value=leaf_value)

    # Create child nodes and recurse
    root = TreeNode(feature_index=best_feature)
    unique_values = data[best_feature].unique()
    
    for value in unique_values:
        subset = data[data[best_feature] == value]
        child_node = create_tree(subset, features, target_feature, max_depth, current_depth + 1)
        root.children[value] = child_node
        
    return root


# --- Example Usage in Jupyter Notebook ---
if __name__ == '__main__':
    features = ['ProductCD', 'card4']
    target = 'isFraud'
    max_tree_depth = 2  # A parameter you would tune

    print("Building a single decision tree...")
    tree = create_tree(data, features, target, max_depth=max_tree_depth)
    
    print("\nPrinting the generated tree:")
    if tree:
        tree.print_tree()
```
